### Without Headers
Your request looks like:

“Hi, I am a script. Definitely not a bot. Trust me.”

Flipkart:

“403. Go away.”

### What is User-Agent?
When you visit a website, your browser sends info like:

Browser name (Chrome, Firefox)

OS (Windows, Mac)
Device type

Example:
Chrome on Windows → Mozilla/5.0 ..

### With headers:

Now your request says:

“Hello, I am Chrome browser on Windows.”

Flipkart:

“Alright, come in.”

```python
# Storing information in lists
titles = []
prices = []
ratings = []
RnR = []
features = []
pgno = []

# loop to go through webpages
for x in range(1,11):
    start_time = time.time()
    
    URL = 'https://www.flipkart.com/search?q=phone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3D20000&p%5B%5D=facets.price_range.to%3DMax&page=1'.format(x)
    
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
    }
    
    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    # Extracting product details of all products on the webpage
    product_details = soup.find_all('div', class_ = 'ZFwe0M row')

    for product in product_details:
        # name
        name = product.find('div', class_ = 'RG5Slk')

        if name is None:
            titles.append(np.nan)
        else:
            titles.append(name.text)

        # price
        price = product.find('div', class_ = 'hZ3P6w DeU9vF')
        if price is None:
            prices.append(np.nan)
        else:
            prices.append(price.text)

        # rating
        rating = product.find('div', class_ = 'MKiFS6')
        if rating is None:
            ratings.append(np.nan)
        else:
            ratings.append(rating.text)

        # ratings and reviews
        ratings_reviews = product.find('span', class_ = 'PvbNMB')
        if ratings_reviews is None:
            RnR.append(np.nan)
        else:
            RnR.append(ratings_reviews.text)

        # specifications
        f = []
        specifications = product.find('div', class_ = 'CMXw7N')
        specs = specifications.find_all('li', class_ = 'DTBslk')
        for feature in specs:
            f.append(feature.text)
        features.append(f)
        
        # page number
        pgno.append(x)
        
    end_time = time.time()
    print('Page {} completed in {} seconds'.format(x, end_time - start_time))
```

```python
print(len(titles))
print(len(prices))
print(len(ratings))
print(len(RnR))
print(len(features))
```

```python
data = {'Product_name' : titles,
        'Price'        : prices,
        'Rating'       : ratings,
        'No_of_RnR'    : RnR,
        'Features'     : features,
        'PageNo'       : pgno}

df = pd.DataFrame(data)
df.head(10)
```

```python
# Converting price and rating to float
df.Rating = df.Rating.astype(float)
# df.Price = df.Price.apply(lambda s : float(s.replace('₹','').replace(',',''))
df.Price = df.Price.str.replace('₹','').str.replace(',','').astype(float)

print(df.Rating.dtype)
print(df.Price.dtype)
df.head()
```

```python
text = '3,346 Ratings & 223 Reviews'
rating_regex = re.compile(r'([\d,\s]+)Ratings')
ratings = rating_regex.findall(text)
int(ratings[0].strip().replace(',',''))
```

```python
# creating functions to get number of ratings and number of reviews
def get_ratings(text):
    if text is np.nan:
        return 0
    else:
        rating_regex = re.compile(r'([\d,\s]+)Ratings')
        ratings = rating_regex.findall(text)
        ratings = int(ratings[0].strip().replace(',',''))
        return ratings
    
def get_reviews(text):
    if text is np.nan:
        return 0
    else:
        review_regex = re.compile(r'([\d,\s]+)Reviews')
        reviews = review_regex.findall(text)
        reviews = int(reviews[0].strip().replace(',',''))
        return reviews
    
print(get_ratings('3,346 Ratings & 223 Reviews'))
print(get_reviews('3,346 Ratings & 223 Reviews'))
```

```python
# Creating columns No_of_ratings and No_of_reviews
df['No_of_ratings'] = df.No_of_RnR.apply(get_ratings)
df['No_of_reviews'] = df.No_of_RnR.apply(get_reviews)
df.drop('No_of_RnR', axis = 1, inplace = True)

df.head(10)
```

```python
# Creating a regex to extract brand
brand_regex = re.compile(r'^\w+')
brand_regex.findall('CMF by Nothing Phone 2 Pro')[0]
```

```python
# creating a brand column
df['Brand'] = df.Product_name.apply(lambda s : brand_regex.findall(s)[0])
df.head(10)
```

```python
# Converting features list to string  (to work with regular expressions)
df.Features = df.Features.apply(lambda x : ','.join(x))
df.head(10)
```

```python
text = df.Features[0]
processor_regex = re.compile(r'([A-Za-z0-9\s]+)Processor')
processor_regex.findall(text)[0].strip()
```

```python
def get_processor(text):
    processor_regex = re.compile(r'([A-Za-z0-9\s]+)Processor')
    processor = processor_regex.findall(text)
    if len(processor) == 0:
        return np.nan
    else:
        return processor[0].strip()
    
get_processor(text)
```

```python
# Creating a Processor column
df['Processor'] = df.Features.apply(get_processor)
df.head(5)
```
